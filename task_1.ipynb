```python
# Compulsory Task 1

# Import packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Read in the data

insurance_df = pd.read_csv("insurance.csv")

# Get a sense of the data
insurance_df.head()
```

```python
# First we look at the relationship between age and insurance charges
# Will use iloc function to determine the columns used

x = insurance_df.iloc[:,0].values.reshape(-1,1)
y = insurance_df.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# Scatter plot to show relationship between age and insurance charges.
# There is a positive correlation. As age increases, insurance charges increase.
```

```python
# Using sklearn

insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
unk_x = [[70]]
age_35 = [[35]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

# The other scatter plot with the line of best fit
plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()

print("Charge for age 70 should be:", insurance_model.predict(unk_x))

print("Charge for age 35 should be:", insurance_model.predict(age_35))

# These predictions are what we would expect from a positive correlation.
```

```python
# Lastly, we will look at the relationship between bmi and insurance charges
# Will use iloc function to determine the columns used

x = insurance_df.iloc[:,2].values.reshape(-1,1)
y = insurance_df.iloc[:,-1].values

plt.scatter(x,y,color = 'g')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()

# Scatter plot to show relationship between age and insurance charges.
# Not a very clear correlation, there appears to be a weak positive correlation. 
# As bmi increases, insurance charges increase.
```
